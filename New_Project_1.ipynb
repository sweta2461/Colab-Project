```python
from PIL import Image

def bmp_to_byte_array(image_path):
    # Open the BMP image
    img = Image.open(image_path)

    # Convert image to grayscale if it's not already
    img = img.convert('L')

    # Convert the image into a byte array
    byte_array = bytearray(img.tobytes())

    return byte_array

# Example usage
image_path = "/content/drive/MyDrive/Images/J-WO-F-bmp-1.bmp"  # Replace with the path to your BMP image
byte_array = bmp_to_byte_array(image_path)
print(byte_array)
```
